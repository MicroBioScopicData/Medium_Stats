## My Medium Statistics
I obtained my Medium statistics for my articles using the Chrome browser extension called "Better Medium Stats" [1]. The resulting CSV file comprises several informative columns:

- **Title**: This column contains the titles of the articles.  

- **Views**: It represents the number of times each article has been viewed.  

- **Reads**: This column indicates the number of times the articles have been read.  

- **R/V (Reads per View)**: This is a calculated column that represents the ratio of Reads to Views. It measures how many reads an article received on average per view.  

- **Claps**: This represents the number of claps given to each article.  

- **Fans**: It indicates the number of fans of the article.  

- **C/F (Claps per Fan)**: Similar to R/V, this column is a calculated ratio of Claps to Fans, showing how many claps an author received on average per fan.  

- **Date**: This column likely indicates the date when each article was published.

```python
#for mathematical computation
import numpy as np
import pandas as pd
#for data visualization
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

#### Dsitribution and outliers
If we investigate the distribution of views, reads, claps,fans, and R/V ratio using box plots with the assistance of the Seaborn library. 

Box plots are effective tools for summarizing the key characteristics of a dataset in a concise and visually informative manner.  

We can mostly observe that: 

- There are some outliers with very high view counts, as shown by the points outside the "whiskers"
- Some articles have significantly more reads than others, as indicated by the outliers

```python
# Create subplots for visualization
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot box plots for Views, Reads, Claps, Fans, and R/V
sns.boxplot(y=df_stories['Views'], ax=axes[1, 2], color='skyblue')
sns.boxplot(y=df_stories['Reads'], ax=axes[0, 0], color='salmon')
sns.boxplot(y=df_stories['Claps'], ax=axes[0, 1], color='lightgreen')
sns.boxplot(y=df_stories['Fans'], ax=axes[0, 2], color='gold')
sns.boxplot(y=df_stories['R/V'], ax=axes[1, 0], color='lightcoral')

# Set titles for the subplots
axes[1, 2].set_title("Distribution of Views")
axes[0, 0].set_title("Distribution of Reads")
axes[0, 1].set_title("Distribution of Claps")
axes[0, 2].set_title("Distribution of Fans")
axes[1, 0].set_title("Distribution of R/V")
```

## Time-Based Analysis:
We can explore how the number of views, reads, and claps have changed over time. Analyzing these numbers can help us understand the impact and engagement of my articles over time. We can observe that:


- In March, which marked the beginning of my writing journey, I experienced the highest levels of views, claps, reads, and fan engagement. This initial surge in activity can be attributed to the support of friends and family who rallied behind my writing efforts. 

- The second month that stands out for its notably high numbers is August, which emerged as my most prolific period. On August we observe an "impressive" 1.07 "R/V" ratio, indicating that readers continued to engage deeply with my content. Moreover, the "C/F" ratio reached the hight value (153.86), implying that fans during this month were extremely generous with their claps.

```python
monthly_stats = df_stories.groupby('Year_Month').agg({
    'Views': 'sum',
    'Claps': 'sum',
    'Reads': 'sum',
    'Fans': 'sum',
    'R/V': 'sum',
    'C/F': 'sum'
}).reset_index()
monthly_stats
```

```python
import matplotlib.pyplot as plt

# Assuming you already have the 'monthly_stats' DataFrame with aggregated data

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot monthly views
ax.plot(monthly_stats['Year_Month'], monthly_stats['Views'], label='Views', marker='o')

# Plot monthly claps
ax.plot(monthly_stats['Year_Month'], monthly_stats['Claps'], label='Claps', marker='o')

# Plot monthly Reads
ax.plot(monthly_stats['Year_Month'], monthly_stats['Reads'], label='Claps', marker='o')

# Customize the plot
ax.set_xlabel('Year-Month')
ax.set_ylabel('Count')
ax.set_title('Monthly Views/Reads and Claps Over Time')
ax.grid(True)
ax.legend()

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

## Correlation Analysis:
We can calculate and visualize correlations between different variables (e.g., views, reads, claps, fans) .
in order to determoine if there is there are any strong relationships between variables. From the data below we can observe that:

- "Views" and "Reads" have a strong positive correlation of approximately 0.96. This indicates that articles with higher views tend to have more reads, suggesting that readers who view an article are likely to read it as well.
- "Views" and "Fans" also have a strong positive correlation of approximately 0.93. This suggests that articles with higher views tend to attract more fans.


```python
# calculate and displays the correlation matrix for the columns 'Views,' 'Reads,' 'Claps,' and 'Fans'
correlation_matrix = df_stories[['Views', 'Reads', 'Claps', 'Fans']].corr()
correlation_matrix
```

```python
# Set the size of the figure 
plt.figure(figsize=(6, 4))
# Create the heatmap using Seaborn 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### Title length

We can observe:

- while there are some very weak correlations between Title_Length and the other variables, the length of article titles does not appear to have a substantial impact on the performance metrics such as Views, Reads, Claps, and Fans. 

```python
# Add a new column called 'Title_Length' to the DataFrame df_stories
df_stories['Title_Length'] = df_stories['Title'].apply(lambda x: len(x))

# Calculate the correlation with 'Reads'
# calculate and displays the correlation matrix for the columns 'Views,' 'Reads,' 'Claps,' and 'Fans'
correlation_matrix_Title_Length = df_stories[['Views', 'Reads', 'Claps', 'Fans', 'Title_Length']].corr()
correlation_matrix_Title_Length
```
